```python
import pandas as pd
```

```python
data = pd.read_csv(r'D:\0_Respaldo\000_KaggleX\Projects\Rag-Ley675\675_LAW\data\clean\articulos.csv')
```

```python
df = data.copy()
```

```python
df.head(1)
```

```python
df['contenido'] = df['contenido'].astype(str)
```

```python
df['conteo_palabras'] = df['contenido'].str.split().apply(len)
```

```python
df['conteo_palabras'].describe()
```

```python
# Definir los rangos para agrupar el conteo de palabras
bins = [0, 132, 216, 910]  # Puedes ajustar estos valores
labels = ['0-132', '131-216', '217-910']
```

```python
# Crear la columna de rangos etáreos
df['rango_palabras'] = pd.cut(df['conteo_palabras'], bins=bins, labels=labels, right=True)
```

```python
df.head(5)
```

```python
# Filtrar las filas que se encuentran en el rango '0-50'
df_rango_0_132 = df[df['rango_palabras'] == '0-132']
```

```python
df_rango_133_216 = df[df['rango_palabras'] == '131-216']
```

```python
df_rango_217_910 = df[df['rango_palabras'] == '131-216']
```
